Split Data-(6840,10)

```python
x=df.drop('mental_fitness',axis=1)
y=df['mental_fitness']

from sklearn.model_selection import train_test_split  #use to split original data into training data & test data
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size=.20, random_state=2)

#random_state simply sets seed to the random generator, so that your train test splits are always dterministic. If you don't set seed,it is different each time.
```

```python
#Training (6840,10)
#6840*80/100= 5472
#6840*20/100= 1368

print("xtrain: ",xtrain.shape)
print("xtest: ",xtest.shape)
print("\n ytrain: ",ytrain.shape)
print("ytest: ",ytest.shape)
```

Model Training

```python
import pandas as pd

# Assuming you have already defined and prepared xtrain and xtest data

# Find non-numeric columns in xtrain
non_numeric_cols_xtrain = xtrain.select_dtypes(exclude=['number'])

# Find non-numeric columns in xtest
non_numeric_cols_xtest = xtest.select_dtypes(exclude=['number'])

print("Non-numeric columns in xtrain:", non_numeric_cols_xtrain)
print("Non-numeric columns in xtest:", non_numeric_cols_xtest)
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming non_numeric_cols_xtrain contains the column names with non-numeric values in xtrain
label_encoder = LabelEncoder()

for col in non_numeric_cols_xtrain:
    xtrain[col] = label_encoder.fit_transform(xtrain[col])

# Do the same for xtest if it contains non-numeric values
for col in non_numeric_cols_xtest:
    xtest[col] = label_encoder.transform(xtest[col])
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming you have already defined and prepared xtrain and ytrain data
# If the data contains non-numeric values, make sure they are preprocessed first.

lr = LinearRegression()
lr.fit(xtrain, ytrain)  # Fit training data

# Model evaluation for training set
ytrain_pred = lr.predict(xtrain)

# Mean Square Error (MSE) is the average of the square of the difference between the observed and predicted values of a variable
mse = mean_squared_error(ytrain, ytrain_pred)

# Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(ytrain, ytrain_pred))

# The coefficient of determination (R-squared)
r2 = r2_score(ytrain, ytrain_pred)

print("The Linear Regression model performance for the training set")
print("-----------------------------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 is {}'.format(r2))
```

Evaluation

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming you have already defined and prepared xtrain, ytrain, xtest, and ytest data

# Check for any non-numeric values in xtrain and xtest
non_numeric_cols_xtrain = xtrain.select_dtypes(include=['object']).columns
non_numeric_cols_xtest = xtest.select_dtypes(include=['object']).columns

if not non_numeric_cols_xtrain.empty or not non_numeric_cols_xtest.empty:
   raise ValueError("Non-numeric values found in the training or test data. Please preprocess the data accordingly.")

# Check for any non-numeric values in ytrain and ytest
if not np.issubdtype(ytrain.dtype, np.number) or not np.issubdtype(ytest.dtype, np.number):
    raise ValueError("Non-numeric values found in the target variable (ytrain or ytest). Please preprocess the target variable accordingly.")

lr = LinearRegression()
lr.fit(xtrain, ytrain)  # Fit training data

# Model evaluation for training set
ytrain_pred = lr.predict(xtrain)

# Mean Square Error (MSE) is the average of the square of the difference between the observed and predicted values of a variable
mse_train = mean_squared_error(ytrain, ytrain_pred)

# Root Mean Square Error (RMSE) for training set
rmse_train = np.sqrt(mse_train)

# The coefficient of determination (R-squared) for training set
r2_train = r2_score(ytrain, ytrain_pred)

# Model evaluation for test set
ytest_pred = lr.predict(xtest)

# Mean Square Error (MSE) for test set
mse_test = mean_squared_error(ytest, ytest_pred)

# Root Mean Square Error (RMSE) for test set
rmse_test = np.sqrt(mse_test)

# The coefficient of determination (R-squared) for test set
r2_test = r2_score(ytest, ytest_pred)

print("The Linear Regression model performance for the training set")
print("-----------------------------------------------------------")
print('MSE is {}'.format(mse_train))
print('RMSE is {}'.format(rmse_train))
print('R2 is {}'.format(r2_train))

print("\n\nThe Linear Regression model performance for the test set")
print("-------------------------------------------------------")
print('MSE is {}'.format(mse_test))
print('RMSE is {}'.format(rmse_test))
print('R2 is {}'.format(r2_test))
```
